```python
import pandas as pd
import matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
df = pd.read_excel("surfaceplot.xlsx", skiprows = 1)
```

```python
df
```

```python
y = df["Operating hour"]
x = df["Operating cost"]
z = df["Profit"]
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('scatter')
plt.show()
```

```python
x, y = np.meshgrid(x, y)
z1 = np.transpose([z] * len(z))
```

```python
x
```

```python
y
```

```python
z1
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z1,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
fig.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z1, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
fig.show()
```
