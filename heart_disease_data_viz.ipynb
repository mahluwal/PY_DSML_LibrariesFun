```python
corr_df = df[['age', 'trestbps', 'chol','thalach','oldpeak']] # leaving catagorical variables.
plt.figure(figsize=(12, 12), dpi=80)
scatter_matrix(corr_df)
plt.savefig("correlation_plot")
plt.show()
```

```python
corr_matrix = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']].corr()
corr_matrix
```

##### PCA for 3 components

```python
pca = PCA(n_components = 3)
PC = pca.fit_transform(corr_df)
PC
```

```python
dataframe = pd.DataFrame(data=PC,columns=['PC1','PC2', "PC3"])
```

```python
plt.scatter(dataframe["PC1"],dataframe["PC2"]) # relation b/w first two principle  components
```

```python
### 3D Visualization b/w all three principle components
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
x = dataframe['PC1']
y = dataframe['PC2']
z = dataframe['PC3']
ax.scatter(x, y, z, color='g', marker='o')  
ax.set_xlabel('1st Principle component')
ax.set_ylabel('2nd Principle Component')
ax.set_zlabel('3rd Principle component')
plt.show()
```
