# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([2, 4, 5])
y = np.array([1, 2.3, 2.4])
```

```python
plt.scatter(x, y)
```

```python
w0, w1 = 0.0, 1.0

lr = 0.001
epochs = 1000
error = []

# run 10000 times
for epoch in range(epochs):
    # initialize to 0 -> cost of epoch, Jw_0, Jw_1
    epoch_cost, cost_w0, cost_w1 = 0, 0, 0
    
    for i in range(len(x)):
        # make prediction
        y_pred = (w0 + w1*x[i])
        # append squared error
        epoch_cost += (y[i] - y_pred)**2 
        
        for j in range(len(x)):
            # partial derivative of w0 and w1 for current row
            partial_wrt_w0 = -2 * (y[j] - (w0 + w1*x[j]))
            partial_wrt_w1 = (-2 * x[j]) * (y[j] - (w0 + w1*x[j]))
            
            # increase cost of coeffs
            cost_w0 += partial_wrt_w0
            cost_w1 += partial_wrt_w1

        # calculate new coeffs
        w0 = w0 - lr * cost_w0
        w1 = w1 - lr * cost_w1
        print("w0", w0)
        print("w1", w1)
        
    # keep track of errors - for visualization purposes
    error.append(epoch_cost)
```

```python
y_pred = []
for xi in x:
    yp = w0+w1*xi
    y_pred.append(yp)
```
